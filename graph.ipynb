```python
with open('test-02.xyz.dump', 'r') as file, open('data_1.csv','w') as data_file_1, open('data_2.csv','w') as data_file_2:
    for line in file:
        line = line.strip().split()
        if len(line) == 7:
            if line[0] == '1':
                data_file_1.write(str(line[1] + ',' + line[4])+ '\n')
            else:
                data_file_2.write(str(line[1] + ',' + line[4]) + '\n')
                                  
                
```
